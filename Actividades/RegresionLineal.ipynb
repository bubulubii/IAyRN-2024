***Descripción de la actividad:***
Entrenamiento de un modelo de regresión lineal.  Este modelo hará la predicción de la calificación de un grupo de estudiantes tomando en cuenta el número de horas dedicadas a estudiar.

```python
# 1. Definimos las librerías que utilizaremos.
import matplotlib.pyplot as plt #para que podamos graficar.
import numpy as np #para generar los arreglos y extraer números aleatorios.
import sklearn.linear_model #para obtener los modelos de Machine Learning.

  #Importamos funciones para trabajar con el modelo.
from sklearn.model_selection import train_test_split #para generar conjuntos de datos.
from sklearn.linear_model import LinearRegression #para el modelo de regresión lineal.
from sklearn.metrics import mean_squared_error #para el cálculo del error cuadrático medio.
```

```python
# 2. Cargamos los datos a usar.
np.random.seed(0)
horas_estudio= np.random.rand(100, 1) * 10
calif= 2*horas_estudio +np.random.randn(100,1)*2
```

```python
# 3. Separamos datos en conjunto de entrenamiento y de prueba.
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calif, test_size= 0.2)
```

```python
# 4. Creamos y entrenamos el modelo de regresión lineal.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

regression_model= LinearRegression()
regression_model.fit(train_data, train_labels)
```

```python
# 5. Establecemos predicciones.
predictions = regression_model.predict(test_data)
```

```python
# 6. Calculamos el error cuadrático medio.
error = mean_squared_error(test_labels, predictions)
print("El error cuadrático medio es de:", error)
```

```python
# 6. Graficamos nuestro modelo de regresión lineal según las predicciones.
plt.scatter(test_data, test_labels, color='blue', label='Datos')
plt.plot(test_data, predictions, color='green', linewidth=2.5, label='Modelo de regresión lineal')
plt.title('Modelo de regresión lineal: Relación horas de estudio-calificación')
plt.xlabel("Horas dedicadas a estudiar")
plt.ylabel('Calificación obtenida')
plt.show()
```
